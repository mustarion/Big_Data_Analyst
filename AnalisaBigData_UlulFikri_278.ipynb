# Assesment

### Data set berikut nantinya dapat digunakan untuk melakukan analisa tipe course yang mana mahasiswa paling banyak mendapat skor tinggi, lalu hasilnya dapat digunakan untuk memberikan pengembangan kepada tipe course lain untuk meningkatkan kualitas dari course yang diberikan. hal tersebut dilakukan dengan cara menyeleksi atau memilih salah satu tahun yaitu 2013 atau 2014 sehingga analisa dapat terfokus pada tahun tersbut saja.

```python
# Melakukan load data assessments lalu menampilkan tabel datanya
data1 = pd.read_csv('assessments.csv')
data1
```

```python
# Menampilkan informasi general pada dataset untuk melihat banyak nilai null serta data type pada masing2 feature
data1.info()
```

```python
# Melihat 5 data teratas pada data
data1.head()
```

```python
# Melihat 5 data paling bawah pada data
data1.tail()
```

```python
# Dilakukan penaganan nilai null pada data berupa drop-null diakrenakan jumlah row yang memiliki nilai null tidak terlalu banyak
data1 = data1.dropna()
data1
```

```python
# Melakukan drop pada tahun 2013j
data1_2014 = data1.loc[data1['code_presentation'] != '2013J']
data1_2014
```

```python
# Melakukan drop pada tahun 2013b
data1_2014 = data1_2014.loc[data1_2014['code_presentation'] != '2013B']
data1_2014
```

```python
# Melihat perbandingan modul yang diberikan pada 2014
tipe_counts = data1_2014['code_module'].value_counts()
plt.figure(figsize=(10, 6)) 


tipe_counts.plot(kind='bar', color='skyblue')

# Menambahkan label sumbu x dan y serta judul
plt.xlabel('Tipe Assigmnet')
plt.ylabel('Jumlah Assignment pada 2014')
plt.title('Perbandingan Jumlah Tipe dalam Kolom')
plt.show()
```

```python
# Melakukan load pada data
# Tidak perlu memperlihatkan head dan tail pada tabelnya karena sudah terlihat semua tabelnya
data2 = pd.read_csv('courses.csv')
data2
```

```python
# Melihat variabel unik pada leght dari fitur module_presentation_length
data2['module_presentation_length'].unique()
```
